### ML Project With MLFLOW

- Installing MLflow.

- Starting a local MLflow Tracking Server.

- Logging and registering a model with MLflow.

- Loading a logged model for inference using MLflow‚Äôs pyfunc flavor.

- Viewing the experiment results in the MLflow UI.

```python
import pandas as pd
from sklearn import datasets
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import train_test_split
```

```python
import mlflow 
from mlflow.models import infer_signature 
```

```python
## set the trackinng ui 
mlflow.set_tracking_uri("http://127.0.0.1:5000")
```

```python
## load the dataset
X,y=datasets.load_iris(return_X_y=True)

# spliting the data into training and test sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)

# Defining the model hyperparameters
params = {"penalty":"l2","solver": "lbfgs", "max_iter": 1000, "multi_class": "auto", "random_state": 8888}

##training the model
lr=LogisticRegression(**params)  ## bcz parameters are in key value pairs
lr.fit(X_train,y_train)
```

```python
print(len(X_test))
X_test
```

```python
## Prediction on the test set
y_pred=lr.predict(X_test)
y_pred
len(y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

## Mlflow Tracking

```python
### MLFLOW tracking
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

##creating a new MLFLOW experiment
mlflow.set_experiment("MLFLOW Quickstart") ## experiment name 


## starting an MLFLOW run

with mlflow.start_run():
    ## log the hyperparameters
    mlflow.log_params(params)

    ## Log the accuracy metrics
    mlflow.log_metric("accuracy",accuracy)

    # Set a tag that we can use to remind ourselves what this run was for
    mlflow.set_tag("Training Info", "Basic LR model for iris data")

    ## Infer the model signature

    signature=infer_signature(X_train,lr.predict(X_test))  ## you need to define input and output

    ## log the model
    model_info=mlflow.sklearn.log_model(
        sk_model=lr,
        artifact_path="iris_model",  ## model name
        signature=signature,
        input_example=X_train,
        registered_model_name="tracking-quickstart",

    )


    
```

```python
# Define the model hyperparameters
params = {"solver": "newton-cg", "max_iter": 1000, "multi_class": "auto", "random_state": 1000}

##train the model

lr=LogisticRegression(**params)
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
## Start an MLFLOW run
with mlflow.start_run():
    ## log the hyperparameters
    mlflow.log_params(params)

    ## Log the accuracy metrics
    mlflow.log_metric("accuracy",accuracy)

    # Set a tag that we can use to remind ourselves what this run was for
    mlflow.set_tag("Training Info", "Basic LR model for iris data")

    ## Infer the model signature

    signature=infer_signature(X_train,lr.predict(X_test))

    ## log the model
    model_info=mlflow.sklearn.log_model(
        sk_model=lr,
        artifact_path="iris_model",
        signature=signature,
        input_example=X_train,
        registered_model_name="tracking-quickstart",

    )
```
